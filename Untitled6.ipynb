```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
import os

folder_path = 'drive/MyDrive/ml/neutral/'

all_files = os.listdir(folder_path)

image_files = [file for file in all_files if file.lower().endswith(('.png'))]

image_dict = {}

for image_file in image_files:
    parts = image_file.split('_')
    prefix = parts[0]
    if prefix not in image_dict:
        image_dict[prefix] = []
    image_dict[prefix].append(image_file)
```

در بالا فایلی که تصاویر را دارد باز کردیم و عکس ها را با توجه به شماره عکس اسمشان را داخل یک دیکشنری که شماره عکس و نام کامل عکس میباشد ریختیم.

```python
import random
all_keys = list(image_dict.keys())

random.shuffle(all_keys)

train_keys = all_keys[:70]
test_keys = all_keys[70:]

train_dict = {key: image_dict[key] for key in train_keys}
test_dict = {key: image_dict[key] for key in test_keys}
```

در قسمت کد بالا با توجه به شماره عکس های موجود 70 عدد آن ها را به صورت رندوم انتخاب کردیم.
ابتدا شماره عکس ها را به صورت رندوم سورت کردیم و سپس 70 تای اول را برداشتیم و باقی مانده را در تست ریختیم.

```python
import pandas as pd
import cv2

train_img = []
test_img = []
path = 'drive/MyDrive/ml/neutral/'
for key in train_dict.keys():
  for img_name in train_dict[key]:
    n= cv2.imread('drive/MyDrive/ml/neutral/'+img_name)
    train_img.append(cv2.cvtColor(n, cv2.COLOR_BGR2GRAY))

for key in test_dict.keys():
  for img_name in test_dict[key]:
    n= cv2.imread('drive/MyDrive/ml/neutral/'+img_name)
    test_img.append(cv2.cvtColor(n, cv2.COLOR_BGR2GRAY))
```
