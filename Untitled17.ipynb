```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # For displaying images in Colab
import os
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import zipfile
# Video file path
video_path = "/content/WIN_20231115_10_06_19_Pro.mp4"

# Read the video
cap = cv2.VideoCapture(video_path)
```

```python
# Dense Optical Flow
ret, frame1 = cap.read()
prvs = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)

hsv = np.zeros_like(frame1)
hsv[..., 1] = 255

for i in range(10):
    ret, frame2 = cap.read()
    if not ret:
        break

    next_img = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    # Calculate dense optical flow\n",
    flow = cv2.calcOpticalFlowFarneback(prvs, next_img, None, 0.5, 3, 15, 3, 5, 1.2, 0)

    # Calculate the angle and magnitude of the flow
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])

    # Set the hue value according to the angle
    hsv[..., 0] = ang * 180 / np.pi / 2

    # Set the value to the magnitude of the flow
    hsv[..., 2] = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)

    # Convert HSV to BGR
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    cv2_imshow(bgr)  # Display the result in Colab



    # Update the previous frame
    prvs = next_img

cv2.destroyAllWindows()

# Uncomment the function you want to use
# sparse_optical_flow(cap)

# Release the video capture object
cap.release()
```

```python

# Function to compute optical flow
def compute_optical_flow(prev_frame, next_frame):
    prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    next_gray = cv2.cvtColor(next_frame, cv2.COLOR_BGR2GRAY)
    flow = cv2.calcOpticalFlowFarneback(prev_gray, next_gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)
    return flow
# Function to compute optical flow and plot histogram
def compute_and_plot_optical_flow(frames):
    # Variables for histogram
    bins = 8
    hist_values = np.zeros(bins)
    for i in range(len(frames)-1):
        # Compute optical flow
        flow = compute_optical_flow(frames[i], frames[i+1])
        # Calculate phase and flatten to 1D array
        phase = np.angle(flow)
        phase_flat = phase.flatten()
        # Compute histogram of the optical flow phase
        hist, bin_edges = np.histogram(phase_flat, bins=bins, range=(-np.pi, np.pi), density=True)
        hist_values += hist
    # Plot the histogram
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    plt.bar(bin_centers, hist_values, width=2 * np.pi / bins, align='center', color='blue')
    plt.title('Optical Flow Phase Histogram')
    plt.xlabel('Phase')
    plt.ylabel('Frequency')
    plt.show()
# Example usage
compute_and_plot_optical_flow(happyim)
```

```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
# Function to compute optical flow
def compute_optical_flow(prev_frame, next_frame):
    prev_gray = cv2.cvtColor(prev_frame, cv2.COLOR_BGR2GRAY)
    next_gray = cv2.cvtColor(next_frame, cv2.COLOR_BGR2GRAY)
    flow = cv2.calcOpticalFlowFarneback(prev_gray, next_gray, None, 0.5, 3, 15, 3, 5, 1.2, 0)
   return flow
# Function to compute optical flow histograms
def compute_optical_flow_histograms(frames):
    bins = 8
    hist_avg = np.zeros(bins)
    hist_max_pool = np.zeros(bins)
    for i in range(len(frames)-1):
        # Compute optical flow
        flow = compute_optical_flow(frames[i], frames[i+1])
        # Calculate phase and flatten to 1D array
        phase = np.angle(flow)
        phase_flat = phase.flatten()\
        # Compute histograms
        hist, _ = np.histogram(phase_flat, bins=bins, range=(-np.pi, np.pi), density=True)
        # Update average histogram
        hist_avg += hist
        # Update max-pooled histogram
       hist_max_pool = np.maximum(hist_max_pool, hist)
    return np.concatenate([hist_avg, hist_max_pool])
# Load data
happy_data = []
surprise_data = []
labels = []
happy_folder = "/content/surprise/happy videos\"
surprise_folder = "/content/surprise/surprise videos\"
for folder_path, label in [(happy_folder, 0), (surprise_folder, 1)]:
    for video_name in os.listdir(folder_path)[:int(0.7*len(os.listdir(folder_path)))]:
        video_path = os.path.join(folder_path, video_name)
        frames = [cv2.imread(os.path.join(video_path, frame_name)) for frame_name in sorted(os.listdir(video_path))]
        descriptor = compute_optical_flow_histograms(frames)
        labels.append(label)
        if label == 0:
            happy_data.append(descriptor)\
        else:
            surprise_data.append(descriptor)
# Concatenate data and labels
X = np.vstack([happy_data, surprise_data])
y = np.array(labels)
# Split the data into training and testing sets\n",
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Train a classifier (SVM in this case)\n",
classifier = SVC()
classifier.fit(X_train, y_train)
# Test the classifier\n",
y_pred = classifier.predict(X_test)
# Calculate accuracy\n",
accuracy = accuracy_score(y_test, y_pred)
print(f\"Classification Accuracy: {accuracy}\")
```
