<div dir=rtl>

این معیار را بین هر دو هیستوگرام موجود به صورت دستی محاسبه کرده و نتایج را در یک جدول ۱۶ * ۱۶ ارائه دهید. برداشت خود از نتایج را توضیح دهید
</div>

با توجه به جدول میبینیم هر چقدر دو هیستوگرام به یکدیگر شبیه تر باشند مقدار این عبارت کمتر خواهد بود و هرچه هیستوگرام دو فصل تفاوت بیشتری داشته باشند این مقدار بیشتر خواهد بود.

```python
# to do
chi_squar=[]
for i in seasons:
  chi_squarei=[]
  for j in seasons:
    trian_i=train.loc[train['season'].isin([i]) ]
    trian_j=train.loc[train['season'].isin([j]) ]
    histi=np.histogram(trian_i['points'],bins=range(100))
    histj=np.histogram(trian_j['points'],bins=range(100))
    diff=(histi[0]-histj[0])**2
    sums=(histi[0]+histj[0])
    where_0 = np.where(sums == 0)
    sums[where_0] = 1
    chi_squarei.append(1/2*sum(np.divide(diff, sums) ))
  chi_squar.append(chi_squarei)
our=pd.DataFrame(chi_squar)
df_laliga=train
our.head(16)
```

پس از ذخیره کردن نوت‌بوک،
سلول زیر را اجرا کنید تا فایل های csv  را به همراه این نوت‌بوک زیپ کند.
</div>

```python
import zipfile
import os


def compress(file_names):
    print("File Paths:")
    print(file_names)

    # Select the compression mode ZIP_DEFLATED for compression
    # or zipfile.ZIP_STORED to just store the file
    compression = zipfile.ZIP_DEFLATED

    # create the zip file first parameter path/name, second mode
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            # Add file to the zip file
            # first parameter file to zip, second filename in zip
            zf.write('./' + file_name, file_name, compress_type=compression)


file_names = [
    file for file in os.listdir()
    if (file.endswith('csv') or file.endswith('ipynb'))
    and file != 'LaLiga_dataset.csv'
]

compress(file_names)
```
