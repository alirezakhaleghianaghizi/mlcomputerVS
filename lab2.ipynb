```python
import os

directory_path = '/content/surprise videos/surprise videos'

image_dict = {}

def extract_third_part(image_name):
    parts = image_name.split('_')
    if len(parts) >= 3:
        parts = parts[2].split('.')
        return int(parts[0])
    else:
        return 0

for folder_name in os.listdir(directory_path):
    folder_path = os.path.join(directory_path, folder_name)


    if os.path.isdir(folder_path):
        image_list = {}
        i=0
        N=len(os.listdir(folder_path))
        sorted_image_names = sorted(os.listdir(folder_path), key=extract_third_part)
        for filename in sorted_image_names:
            if filename.endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp')):
                #print(filename,(i)/N)
                i=i+1
                image_list[os.path.join(folder_path, filename)]=(i)/N
                #if i+1==N:
                  #image_list[os.path.join(folder_path, filename)]=(i+1)/N

        image_dict[folder_name] = image_list

print(image_dict)


# Sort the image names based on the third part

```

```python
import random
import cv2

dict_items = list(image_dict.items())

# Shuffle the list of items randomly
random.shuffle(dict_items)

# Calculate the number of items to collect (60% of the total)
percent_to_collect = 0.60
num_to_collect = int(percent_to_collect * len(dict_items))

# Select the first 60% of the shuffled items
train_dict = dict(dict_items[:num_to_collect])
test_dict = dict(dict_items[num_to_collect:len(dict_items)])
X_train=[]
y_train=[]
X_test=[]
y_test=[]
for key in train_dict.keys():
  folder1=train_dict[key]
  for sub_key in folder1.keys():
    y_train.append(folder1[sub_key])
    image = cv2.imread(sub_key)
    image =cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_train.append(image)

for key in test_dict.keys():
  folder1=test_dict[key]
  for sub_key in folder1.keys():
    y_test.append(folder1[sub_key])
    image = cv2.imread(sub_key)
    image =cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_test.append(image)
```

```python

X_train=np.array(X_train)
X_train=X_train.reshape(len(X_train),-1)
X_train.shape
X_test=np.array(X_test)
X_test=X_test.reshape(len(X_test),-1)
X_test.shape
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse_test = mean_squared_error(y_test, y_pred)
print(mse_test)
y_pred = model.predict(X_train)
mse_train = mean_squared_error(y_train, y_pred)
print(mse_train)
```
